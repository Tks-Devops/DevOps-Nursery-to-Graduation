```python
MASTERING_GIT
🚀 DevOps Nursery to Graduation – Git Mastery Guide
👨💻 Author: TK Sahu
This repository demonstrates end-to-end Git workflow, including:
•	Local repository setup
•	Remote repository integration
•	Branching strategy
•	Rollbacks & recovery
•	SSH authentication
•	Semantic Versioning
________________________________________
📁 1. Local Repository Setup
Step 1: Create a directory
mkdir mastering_git
cd mastering_git
Step 2: Initialize Git
git init
✅ Output (real terminal)
Initialized empty Git repository in E:/TK SAHU/mastering_git/.git/
Step 3: Check hidden .git folder
ls -a
./  ../  .git/
________________________________________
📂 2. Creating Files & Directories
mkdir earth jupiter pluto
touch moon{1..10}.py
👉 Git does not track empty directories, so we add files:
touch earth/test1
touch jupiter/tester32
touch pluto/tester89
________________________________________
📊 3. Git Status
git status
Output:
Untracked files:
  earth/
  jupiter/
  moon1.py ...
________________________________________
📌 4. Staging & Commit
git add .
git commit -m "new files committed"
Output:
13 files changed
create mode 100644 earth/test1
________________________________________
🌐 5. Remote Repository Setup
Add remote
git remote add origin https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
Rename branch
git branch -M main
________________________________________
❌ Common Error (Authentication)
remote: Invalid username or token
👉 Reason: GitHub no longer supports password authentication.
________________________________________
✅ Fix
Option 1: Use Token
Option 2: Use SSH (Recommended)
________________________________________
🔄 6. Sync with Remote
git pull origin main --rebase
git push -u origin main
________________________________________
✏️ 7. Modify File Example
vim moon1.py
Content:
moon has many more mines
gold
silver
iron
git add moon1.py
git commit -m "moon1 mines details"
git push
________________________________________
📜 8. Git Logs
git log --oneline
2051ca2 moon1 mines details
bd336a6 new files committed
________________________________________
🔍 9. View Changes
git show 2051ca2
________________________________________
🔄 10. Pull Latest Changes
git pull
________________________________________
🌿 11. Branching Strategy
Create branch
git branch sprint1
git checkout sprint1
Make changes
git rm moon6.py moon7.py
git mv moon1.py moon11.py
Commit & Push
git add .
git commit -m "sprint1 changes"
git push origin sprint1
________________________________________
🔀 12. Merge Branch
git switch main
git merge sprint1
________________________________________
📥 13. Clone Repository
git clone https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
________________________________________
🔁 14. Rollback Techniques
1. Undo uncommitted changes
git restore jupiter1.rb
________________________________________
2. Compare changes
git diff
git diff --cached
________________________________________
3. Unstage file
git restore --staged jupiter1.rb
________________________________________
4. Revert committed change (safe)
git revert HEAD
👉 Creates a new commit that undoes previous changes
________________________________________
5. Hard Reset (dangerous)
git reset --hard <commit-id>
⚠️ Deletes history (use carefully)
________________________________________
🔐 15. SSH Authentication (Recommended)
Generate SSH key
ssh-keygen -t ed25519 -C "your_email@example.com"
Check keys
ls ~/.ssh/
id_ed25519
id_ed25519.pub
Copy public key
cat ~/.ssh/id_ed25519.pub
________________________________________
Add to GitHub
➡️ Settings → SSH & GPG Keys → New SSH Key
________________________________________
📦 16. Semantic Versioning (SemVer)
Format
MAJOR.MINOR.PATCH
Example:
2.5.1
________________________________________
🔴 MAJOR
Breaking changes
👉 Example: API change
________________________________________
🟡 MINOR
New feature (backward compatible)
________________________________________
🟢 PATCH
Bug fixes
________________________________________
📌 Example Flow
1.0.0 → Initial release
1.0.1 → Bug fix
1.1.0 → Feature added
2.0.0 → Breaking change
________________________________________
🏷️ Git Tags
git tag v1.0.0
git push origin --tags
________________________________________
🌍 Real-Life Example
Think of an e-commerce app:
•	1.0.0 → Basic shopping
•	1.1.0 → Add payment gateway
•	1.1.1 → Fix payment bug
•	2.0.0 → New UI (breaking change)
________________________________________
🧠 Key Takeaways
•	Git tracks changes, not files
•	Branching = safe experimentation
•	Revert = safe undo
•	Reset = risky undo
•	SSH = industry standard authentication
________________________________________
⭐ Bonus (Git Best Practices)
•	Use meaningful commit messages
•	Always pull before push
•	Use branches for features
•	Avoid --hard reset on shared branches
________________________________________
If you want next level upgrade, I can help you:
👉 Convert this into portfolio project (DevOps Alchemy ready)
👉 Add GitHub diagrams + CI/CD integration
👉 Add interview questions + answers section
Just tell me 👍
The section covers the following topics: What is a VCS and Why VCS?: The fundamentals of version control.Centralized vs. Distributed VCS: Comparing models like Subversion (Centralized) and Git (Distributed). What is Git?: An introduction to the specific tool.Git repositories: Understanding where code is stored.Versioning with Git: How to track changes.GitHub: Using the popular cloud-based hosting service for Git.Git concepts: Core mechanics like branching and merging.Lots of Git Commands: Practical command-line usage.GitHub SSH Login: Setting up secure access.GitHub Copilot (AI): Leveraging AI to assist with coding and DevOps tasks.
Here’s a clean, professional README section for your content. I’ve structured it so it looks strong in GitHub and is easy for recruiters/interviewers to scan 👇
________________________________________
📘 Git & Version Control – Complete Learning Path
This section covers the fundamentals of Version Control Systems (VCS) and practical usage of Git and GitHub in real-world DevOps environments.
________________________________________
🔹 What is a VCS and Why VCS?
A Version Control System (VCS) helps track changes in code over time.
✅ Why we need VCS:
•	Track history of changes
•	Collaborate with teams
•	Rollback to previous versions
•	Avoid overwriting others’ work
🌍 Real-life Example:
Think of Google Docs:
•	Multiple people edit the same file
•	You can view history and restore older versions
👉 Git works in a similar way for code.
________________________________________
🔹 Centralized vs Distributed VCS
🏢 Centralized VCS (e.g., SVN)
•	Single central server
•	Developers commit to one location
❌ Disadvantages:
•	Single point of failure
•	Requires internet connection
________________________________________
🌐 Distributed VCS (e.g., Git)
•	Each developer has a full copy of the repository
•	Works offline
✅ Advantages:
•	Faster operations
•	Better collaboration
•	No dependency on central server
________________________________________
🔹 What is Git?
Git is a distributed version control system used to:
•	Track changes in source code
•	Manage collaboration between developers
•	Maintain project history
________________________________________
🔹 Git Repositories
A repository (repo) is a storage location for your project.
Types:
•	Local Repo → On your machine
•	Remote Repo → Hosted on platforms like GitHub
________________________________________
🔹 Versioning with Git
Git tracks changes using commits.
Example:
git add .
git commit -m "Added login feature"
👉 Each commit = snapshot of your project
________________________________________
🔹 GitHub
GitHub is a cloud platform to host Git repositories.
Features:
•	Code hosting
•	Collaboration
•	Pull requests
•	CI/CD integrations
________________________________________
🔹 Core Git Concepts
🌿 Branching
•	Create separate environments for features
git branch feature1
git checkout feature1
________________________________________
🔀 Merging
•	Combine changes into main branch
git checkout main
git merge feature1
________________________________________
🔹 Common Git Commands
Command	Purpose
git init	Initialize repo
git clone	Copy repo
git status	Check changes
git add	Stage changes
git commit	Save changes
git push	Upload to remote
git pull	Fetch latest changes
________________________________________
🔹 GitHub SSH Login
Why SSH?
•	More secure than password/token
•	No need to enter credentials repeatedly
Setup:
ssh-keygen -t ed25519 -C "your_email@example.com"
cat ~/.ssh/id_ed25519.pub
👉 Add key to GitHub → Settings → SSH Keys
________________________________________
🔹 GitHub Copilot (AI)
GitHub Copilot is an AI tool that helps developers write code faster.
💡 Uses:
•	Auto-generate code
•	Suggest scripts
•	Help in DevOps automation
Example:
You type:
# create a dockerfile for python app

LAB

Creating a folder inside the e-drive …name “mastering_git”
cmd--$ mkdir mastering_git
$ cd mastering_git
then initialize the git using cmd
$ git init
then check the .git hidden folder where every configurion and versioning will avelible
$ ls –a

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git
$ git init
Initialized empty Git repository in E:/TK SAHU/mastering_git/.git/

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ ls

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ ls -a
./  ../  .git/

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$

Nest created some dir & files inside the masterging git folder


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)

$ mkdir earth jupiter pluto

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)

$ touch moon{1..10}.py

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ ls
earth/  jupiter/  moon1.py  moon10.py  moon2.py  moon3.py  moon4.py  moon5.py  moon6.py  moon7.py  moon8.py  moon9.py  pluto/

create some fles inside the all dir ..bcoz git will not track the empty dir

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ touch earth/test1

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ touch jupiter/tester32

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ touch pluto/tester89

Now check the status using cmd – git status ---here this cmd will show all untracked file

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git status
On branch master

No commits yet

Untracked files:
  (use "git add <file>..." to include in what will be committed)
        earth/
        jupiter/
        moon1.py
        moon10.py
        moon2.py
        moon3.py
        moon4.py
        moon5.py
        moon6.py
        moon7.py
        moon8.py
        moon9.py
        pluto/

nothing added to commit but untracked files present (use "git add" to track)


now we will do stage the file using cmd git add . (whatever untracked file is present that will tracked )

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git add .

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git status
On branch master

No commits yet

Changes to be committed:
  (use "git rm --cached <file>..." to unstage)
        new file:   earth/test1
        new file:   jupiter/tester32
        new file:   moon1.py
        new file:   moon10.py
        new file:   moon2.py
        new file:   moon3.py
        new file:   moon4.py
        new file:   moon5.py
        new file:   moon6.py
        new file:   moon7.py
        new file:   moon8.py
        new file:   moon9.py
        new file:   pluto/tester89

now commit the file

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git commit -m "new files commited"
[master (root-commit) 53475de] new files commited
 13 files changed, 0 insertions(+), 0 deletions(-)
 create mode 100644 earth/test1
 create mode 100644 jupiter/tester32
 create mode 100644 moon1.py
 create mode 100644 moon10.py
 create mode 100644 moon2.py
 create mode 100644 moon3.py
 create mode 100644 moon4.py
 create mode 100644 moon5.py
 create mode 100644 moon6.py
 create mode 100644 moon7.py
 create mode 100644 moon8.py
 create mode 100644 moon9.py
 create mode 100644 pluto/tester89

this cmd will  help to see any

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git status
On branch master
nothing to commit, working tree clean



Local Repository Setup
1.	Set the email for git to use when you commit:
                 Git config –global user.name “tk sahu”
                 Git config –global user.email tksahu@gmail.com
2.	Create a directory
3.	Initialize dir with
    Git init
4.	Create Readme.md file
	Git add (staging)
	Git commit (Local commit)

Remote Repository

1.	Cretae Remote repository on
	Github,bitbucket,codecommit etc
2.	Clone Repo to Local
	Git clone URL
3.	Local to Remote integration
	Cd to local repo
	Git remote add origin
Ssh://git@github.com/(username)/(repo-name).git
	Git push
	Git pull (to fetch latest changes)


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git remote add origin https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ cat .git/config
[core]
        repositoryformatversion = 0
        filemode = false
        bare = false
        logallrefupdates = true
        symlinks = false
        ignorecase = true
[remote "origin"]
        url = https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
        fetch = +refs/heads/*:refs/remotes/origin/*

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (master)
$ git branch -M main


After this configuration we need to push the code to remote …while pushing the code we need to provide github tocken as  well as we need to use “git pull origin main --rebase”

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git push -u origin main
remote: Invalid username or token. Password authentication is not supported for Git operations.
fatal: Authentication failed for 'https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git/'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git push -u origin main
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
 ! [rejected]        main -> main (fetch first)
error: failed to push some refs to 'https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git'
hint: Updates were rejected because the remote contains work that you do not
hint: have locally. This is usually caused by another repository pushing to
hint: the same ref. If you want to integrate the remote changes, use
hint: 'git pull' before pushing again.
hint: See the 'Note about fast-forwards' in 'git push --help' for details.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git pull origin main --rebase
remote: Enumerating objects: 661, done.
remote: Counting objects: 100% (7/7), done.
remote: Compressing objects: 100% (6/6), done.
remote: Total 661 (delta 5), reused 1 (delta 1), pack-reused 654 (from 3)
Receiving objects: 100% (661/661), 94.00 MiB | 3.83 MiB/s, done.
Resolving deltas: 100% (228/228), done.
From https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation
 * branch            main       -> FETCH_HEAD
 * [new branch]      main       -> origin/main
Successfully rebased and updated refs/heads/main.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git push -u origin main
Enumerating objects: 7, done.
Counting objects: 100% (7/7), done.
Delta compression using up to 4 threads
Compressing objects: 100% (2/2), done.
Writing objects: 100% (6/6), 773 bytes | 48.00 KiB/s, done.
Total 6 (delta 0), reused 1 (delta 0), pack-reused 0 (from 0)
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
   dc5cc6e..bd336a6  main -> main
branch 'main' set up to track 'origin/main'.



Lets add some data inside the moon1.py folder and stage the file and push to the remote repo

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ vim moon1.py

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ cat moon1.py
moon has many more mines
gold
silver
iron:

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git status
On branch main
Your branch is up to date with 'origin/main'.

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git restore <file>..." to discard changes in working directory)
        modified:   moon1.py

no changes added to commit (use "git add" and/or "git commit -a")

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git add moon1.py
warning: in the working copy of 'moon1.py', LF will be replaced by CRLF the next time Git touches it

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git commit -m "moon1 mines details"
[main 2051ca2] moon1 mines details
 1 file changed, 4 insertions(+)

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git push origin main
Enumerating objects: 5, done.
Counting objects: 100% (5/5), done.
Delta compression using up to 4 threads
Compressing objects: 100% (2/2), done.
Writing objects: 100% (3/3), 304 bytes | 152.00 KiB/s, done.
Total 3 (delta 1), reused 0 (delta 0), pack-reused 0 (from 0)
remote: Resolving deltas: 100% (1/1), completed with 1 local object.
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
   bd336a6..2051ca2  main -> main

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$



Git log cmd will show you all the previous cmd of current branch

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git log
commit 2051ca2119d2bbd218e451f4cdeb4e15df584a07 (HEAD -> main, origin/main)
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 18:00:08 2026 +0530

    moon1 mines details

commit bd336a6457f61f62aee334122727c0ea53f75286
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 17:06:00 2026 +0530

    new files commited

commit dc5cc6ea946b2a7f847b4204fbe224b105443cca
Author: Tks DevOps <tksahu@example.com>
Date:   Fri Feb 21 00:51:30 2025 +0530

    Added Petclinic folder

commit 975ac0be9aa58902e59ddfd80e5d067596d7e77c
Author: Tulasi kumar sahu <tksahu.devops@gmail.com>
Date:   Thu Feb 20 10:12:42 2025 +0530

    Update MAVEN.md

commit de8d755aece16ffcfe8cad5985eb86a2069e43a6
Author: Tulasi kumar sahu <tksahu.devops@gmail.com>
Date:   Thu Feb 20 10:09:40 2025 +0530

    Update MAVEN.md

commit c9d60769828c4cfa705cd2dac85855c962a33478
Author: Tulasi kumar sahu <tksahu.devops@gmail.com>
Date:   Thu Feb 20 10:03:14 2025 +0530

    Create MAVEN.md

commit c020f622b2290a44b83d3ff28c212259fe6dffef
Author: tksahu <tksahu.devops@gmail.com>
Date:   Thu Feb 20 09:45:08 2025 +0530

Above one gives you huge dtails so if we want minimum information we can use this cmd

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git log --oneline
2051ca2 (HEAD -> main, origin/main) moon1 mines details
bd336a6 new files commited
dc5cc6e Added Petclinic folder
975ac0b Update MAVEN.md
de8d755 Update MAVEN.md
c9d6076 Create MAVEN.md
c020f62 Added springboot-java-project properly
e13f9f6 Added springboot-java-project
b7c04b0 Update README.md
fdce5fd Update README.md
254233c Update README.md
b75fcc5 Add files via upload
924d0f4 Add files via upload
4d8c58c Update README.md
6f8d882 Update README.md
374cdc3 Update README.md
c2e9f26 Update README.md
6453020 Update README.md
9bac826 Update README.md
e191825 Initial commit


now show the what code is return inside the file “moon1.py” using cmd line ‘git show hash-id/commit id”

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git show 2051ca2
commit 2051ca2119d2bbd218e451f4cdeb4e15df584a07 (HEAD -> main, origin/main)
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 18:00:08 2026 +0530

    moon1 mines details

diff --git a/moon1.py b/moon1.py
index e69de29..c64a51f 100644
--- a/moon1.py
+++ b/moon1.py
@@ -0,0 +1,4 @@
+moon has many more mines
+gold
+silver
+iron:

Lets few more change in the moon1.py file

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ vim moon1.py

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ cat moon1.py
moon has many more mines
gold
silver
iron
uranium
boxaite
coal


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git add moon1.py
warning: in the working copy of 'moon1.py', LF will be replaced by CRLF the next time Git touches it

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git commit -m "updated moons mines name"
[main 6bffce5] updated moons mines name
 1 file changed, 5 insertions(+), 1 deletion(-)

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git push origin main
Enumerating objects: 5, done.
Counting objects: 100% (5/5), done.
Delta compression using up to 4 threads
Compressing objects: 100% (3/3), done.
Writing objects: 100% (3/3), 321 bytes | 321.00 KiB/s, done.
Total 3 (delta 1), reused 0 (delta 0), pack-reused 0 (from 0)
remote: Resolving deltas: 100% (1/1), completed with 1 local object.
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
   2051ca2..6bffce5  main -> main


After edit ya add some file you can observe like some line will add or remove


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git log --oneline
6bffce5 (HEAD -> main, origin/main) updated moons mines name
2051ca2 moon1 mines details
bd336a6 new files commited
dc5cc6e Added Petclinic folder
975ac0b Update MAVEN.md
de8d755 Update MAVEN.md
c9d6076 Create MAVEN.md
c020f62 Added springboot-java-project properly
e13f9f6 Added springboot-java-project
b7c04b0 Update README.md
fdce5fd Update README.md
254233c Update README.md
b75fcc5 Add files via upload
924d0f4 Add files via upload
4d8c58c Update README.md
6f8d882 Update README.md
374cdc3 Update README.md
c2e9f26 Update README.md
6453020 Update README.md
9bac826 Update README.md
e191825 Initial commit

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git show 6bffce5
commit 6bffce556c64a9fc5ffc1e605a6b579277636b5e (HEAD -> main, origin/main)
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 18:30:06 2026 +0530

    updated moons mines name

diff --git a/moon1.py b/moon1.py
index c64a51f..f34a5c8 100644
--- a/moon1.py
+++ b/moon1.py
@@ -1,4 +1,8 @@
 moon has many more mines
 gold
 silver
-iron:
+iron
+uranium
+boxaite
+coal
+



Next task some one is push some new changes in the remote repo lets see which one not change by us  ..i have not the change ..other contributer made the change ..using “git pull” cmd to update the change and check “git log” to see the changes ..


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git pull
remote: Enumerating objects: 31, done.
remote: Counting objects: 100% (31/31), done.
remote: Compressing objects: 100% (23/23), done.
remote: Total 28 (delta 8), reused 14 (delta 2), pack-reused 0 (from 0)
Unpacking objects: 100% (28/28), 6.52 KiB | 2.00 KiB/s, done.
From https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation
   6bffce5..06128c0  main                  -> origin/main
 * [new branch]      feature-collaboration -> origin/feature-collaboration
 * [new branch]      master                -> origin/master
 * [new tag]         v1.0                  -> v1.0
Updating 6bffce5..06128c0
Fast-forward
 moon10.py | 4 ++++
 1 file changed, 4 insertions(+)

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git log
commit 06128c07fc913a01d09101791afbf6eb69884e61 (HEAD -> main, origin/main, origin/HEAD)
Author: Tulasi kumar sahu <tksahu.devops@gmail.com>
Date:   Thu Apr 30 18:53:18 2026 +0530

    Add resource descriptions to moon10.py

    updated moons  mines details

commit 6bffce556c64a9fc5ffc1e605a6b579277636b5e
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 18:30:06 2026 +0530

    updated moons mines name

commit 2051ca2119d2bbd218e451f4cdeb4e15df584a07
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 18:00:08 2026 +0530

    moon1 mines details

commit bd336a6457f61f62aee334122727c0ea53f75286
Author: tk <tulasisahu.sahu9@gmail.com>
Date:   Thu Apr 30 17:06:00 2026 +0530

    new files commited




**************BRANCHES & MORE ******************


Every repo having a main branch which was used for stable code ..so any thing need to change ..done by developer or for a period of time ..like an agile environment we have sprints, dfferent iteration …in iteration one ---sprint one may be developed some set of features and sprint two you will have developed some features.

For that period of time you can create a new branch which was copy of the existing branch ..you can keep making changes the branch but main branch will be intact ..so the main branch will have stable copy..
And once all the changes are done or everything is stable on the other sprint branch you can merge it with the main branch.

1.	Lets create a branch “git branch –c sprint1” after checkout the branch you will observe exart copy of main branch ..exartly same data..

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git branch -c sprint1

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git branch -a
* main
  sprint1
  remotes/origin/HEAD -> origin/main
  remotes/origin/feature-collaboration
  remotes/origin/main
  remotes/origin/master

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git checkout sprint1
Switched to branch 'sprint1'
Your branch is up to date with 'origin/main'.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     moon1.py    moon2.py   moon4.py   moon6.py   moon8.py   pluto/         springboot-java-poject/
 MAVEN.md                                                      README.md    jupiter/   moon10.py   moon3.py   moon5.py   moon7.py   moon9.py  'python .jpg'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$
2.	Lets do some drastic change in the sprint1 branch ..delete some file inside the branch

Git rm moon6.py moon7.py moon8.py moon9.py
TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git rm moon6.py moon7.py moon8.py moon9.py
rm 'moon6.py'
rm 'moon7.py'
rm 'moon8.py'
rm 'moon9.py'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     moon1.py    moon2.py   moon4.py   pluto/         springboot-java-poject/
 MAVEN.md                                                      README.md    jupiter/   moon10.py   moon3.py   moon5.py  'python .jpg'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git status
On branch sprint1
Your branch is up to date with 'origin/main'.

Changes to be committed:
  (use "git restore --staged <file>..." to unstage)
        deleted:    moon6.py
        deleted:    moon7.py
        deleted:    moon8.py
        deleted:    moon9.py


3.	Copy some file , delete some file and add some new file ..push to remote.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git mv moon1.py moon11.py

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git rm moon2.py moon3.py moon4.py moon5.py
rm 'moon2.py'
rm 'moon3.py'
rm 'moon4.py'
rm 'moon5.py'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   MAVEN.md   Petclinic/   README.md   earth/   jupiter/   moon10.py   moon11.py   pluto/  'python .jpg'   springboot-java-poject/

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ touch jupiter{1..4}.rb

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     jupiter1.rb   jupiter3.rb   moon10.py   pluto/         springboot-java-poject/
 MAVEN.md                                                      README.md    jupiter/   jupiter2.rb   jupiter4.rb   moon11.py  'python .jpg'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git add .

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git commit -m "jupiter changes "
[sprint1 432e995] jupiter changes
 9 files changed, 0 insertions(+), 0 deletions(-)
 rename moon2.py => jupiter1.rb (100%)
 rename moon3.py => jupiter2.rb (100%)
 rename moon4.py => jupiter3.rb (100%)
 rename moon5.py => jupiter4.rb (100%)
 rename moon1.py => moon11.py (100%)
 delete mode 100644 moon6.py
 delete mode 100644 moon7.py
 delete mode 100644 moon8.py
 delete mode 100644 moon9.py

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git push origin sprint1
Enumerating objects: 3, done.
Counting objects: 100% (3/3), done.
Delta compression using up to 4 threads
Compressing objects: 100% (2/2), done.
Writing objects: 100% (2/2), 264 bytes | 264.00 KiB/s, done.
Total 2 (delta 1), reused 0 (delta 0), pack-reused 0 (from 0)
remote: Resolving deltas: 100% (1/1), completed with 1 local object.
remote:
remote: Create a pull request for 'sprint1' on GitHub by visiting:
remote:      https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation/pull/new/sprint1
remote:
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
 * [new branch]      sprint1 -> sprint1

4.	We can find out new branch in the github folder



Again one more developer will add some features  so he created branch under the sprint1 branch and name as sprint2 ..he add some file push to remote ..for that we use “git pull”

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git pull
From https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation
 * [new branch]      sprint2    -> origin/sprint2
Already up to date.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git branch
  main
* sprint1

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git checkout sprint2
branch 'sprint2' set up to track 'origin/sprint2'.
Switched to a new branch 'sprint2'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     jupiter1.rb   jupiter3.rb   moon10.py   pluto/         springboot-java-poject/
 MAVEN.md                                                      README.md    jupiter/   jupiter2.rb   jupiter4.rb   moon11.py  'python .jpg'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git branch
  main
  sprint1
* sprint2

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)



5.	In “sprint2” branch again add some new file

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ touch sun earth mercury venus

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     jupiter1.rb   jupiter3.rb   mercury     moon11.py  'python .jpg'              sun
 MAVEN.md                                                      README.md    jupiter/   jupiter2.rb   jupiter4.rb   moon10.py   pluto/      springboot-java-poject/   venus

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git add .

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git commit -m "planets & stars"
[sprint2 4a71924] planets & stars
 3 files changed, 0 insertions(+), 0 deletions(-)
 create mode 100644 mercury
 create mode 100644 sun
 create mode 100644 venus

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git push origin sprint2
Enumerating objects: 3, done.
Counting objects: 100% (3/3), done.
Delta compression using up to 4 threads
Compressing objects: 100% (2/2), done.
Writing objects: 100% (2/2), 282 bytes | 141.00 KiB/s, done.
Total 2 (delta 1), reused 0 (delta 0), pack-reused 0 (from 0)
remote: Resolving deltas: 100% (1/1), completed with 1 local object.
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
   432e995..4a71924  sprint2 -> sprint2

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git branch -a
  main
  sprint1
* sprint2
  remotes/origin/HEAD -> origin/main
  remotes/origin/feature-collaboration
  remotes/origin/main
  remotes/origin/master
  remotes/origin/sprint1
  remotes/origin/sprint2



6.	Lets switch from sprint2 branch to sprint1 then again checkout to main bcoz we merge the sprint1 branch to main branch .


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint2)
$ git switch sprint1
Switched to branch 'sprint1'
Your branch is ahead of 'origin/main' by 1 commit.
  (use "git push" to publish your local commits)

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ ls
'Blue And Green Professional Technology LinkedIn Banner.png'   Petclinic/   earth/     jupiter1.rb   jupiter3.rb   moon10.py   pluto/         springboot-java-poject/
 MAVEN.md                                                      README.md    jupiter/   jupiter2.rb   jupiter4.rb   moon11.py  'python .jpg'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git switch main
Switched to branch 'main'
Your branch is up to date with 'origin/main'.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (main)
$ git merge sprint1
Updating 06128c0..432e995
Fast-forward
 moon2.py => jupiter1.rb | 0
 moon3.py => jupiter2.rb | 0
 moon4.py => jupiter3.rb | 0
 moon5.py => jupiter4.rb | 0
 moon1.py => moon11.py   | 0
 moon6.py                | 0
 moon7.py                | 0
 moon8.py                | 0
 moon9.py                | 0
 9 files changed, 0 insertions(+), 0 deletions(-)
 rename moon2.py => jupiter1.rb (100%)
 rename moon3.py => jupiter2.rb (100%)
 rename moon4.py => jupiter3.rb (100%)
 rename moon5.py => jupiter4.rb (100%)
 rename moon1.py => moon11.py (100%)
 delete mode 100644 moon6.py
 delete mode 100644 moon7.py
 delete mode 100644 moon8.py
 delete mode 100644 moon9.py


7.	Now push all braches code to remote and then delete the “mastering_git” folder for prcaicing some new command.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ git push --all origin
Total 0 (delta 0), reused 0 (delta 0), pack-reused 0 (from 0)
To https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
   06128c0..432e995  main -> main

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/mastering_git (sprint1)
$ cd ..

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU
$ rm -rf mastering_git

The new command is “git clone https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git” this is a public repo so not asking any password and token ..if this repo will private then we should provide the keys ya token..





**************RollBack ******************


1.	Lets add some code inside the “jupiter1.rb” file and again try to roll back it .
For rollback I will use the cmd is “git checkout (filename)”

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ vim jupiter1.rb

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ cat jupiter1.rb
- name: Setup Tomcat8 & Deploy Artifact
  hosts: appsrvgrp
  # gather_facts: no
  vars:
    timestamp: "{{ansible_date_time.date}}{{ansible_date_time.hour}}{{ansible_date_time.minute}}"
    tom_url: https://archive.apache.org/dist/tomcat/tomcat-8/v8.5.37/bin/apache-tomcat-8.5.37.tar.gz
  tasks:
    - name: Install JDK on Centos 6/7
      yum:
        name: java-1.8.0-openjdk.x86_64
        state: present
      when: ansible_distribution == 'CentOS'

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git checkout jupiter1.rb
Updated 1 path from the index

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ cat jupiter1.rb

2.	Here some cmd like “git diff” and “git status”.. “git diff” will help to show the comparison between two same two file ya different file.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ vim jupiter1.rb

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git status
On branch main
Your branch is up to date with 'origin/main'.

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git restore <file>..." to discard changes in working directory)
        modified:   jupiter1.rb

no changes added to commit (use "git add" and/or "git commit -a")

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff
warning: in the working copy of 'jupiter1.rb', LF will be replaced by CRLF the next time Git touches it
diff --git a/jupiter1.rb b/jupiter1.rb
index e69de29..3e27566 100644
--- a/jupiter1.rb
+++ b/jupiter1.rb
@@ -0,0 +1,7 @@
+- name: Setup Tomcat8 & Deploy Artifact
+  hosts: appsrvgrp
+  # gather_facts: no
+  vars:
+    timestamp: "{{ansible_date_time.date}}{{ansible_date_time.hour}}{{ansible_date_time.minute}}"
+    tom_url: https://archive.apache.org/dist/tomcat/tomcat-8/v8.5.37/bin/apache-tomcat-8.5.37.tar.gz
+

3.	If you want to see the different of staged file then we need use “git diff –cached” cmd

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git add .
warning: in the working copy of 'jupiter1.rb', LF will be replaced by CRLF the next time Git touches it

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git status
On branch main
Your branch is up to date with 'origin/main'.

Changes to be committed:
  (use "git restore --staged <file>..." to unstage)
        modified:   jupiter1.rb


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff --cached
diff --git a/jupiter1.rb b/jupiter1.rb
index e69de29..3e27566 100644
--- a/jupiter1.rb
+++ b/jupiter1.rb
@@ -0,0 +1,7 @@
+- name: Setup Tomcat8 & Deploy Artifact
+  hosts: appsrvgrp
+  # gather_facts: no
+  vars:
+    timestamp: "{{ansible_date_time.date}}{{ansible_date_time.hour}}{{ansible_date_time.minute}}"
+    tom_url: https://archive.apache.org/dist/tomcat/tomcat-8/v8.5.37/bin/apache-tomcat-8.5.37.tar.gz
+

4.	If you want to unstaged the fle then “git restore - -staged (file name) ”

$ git status
On branch main
Your branch is up to date with 'origin/main'.

Changes to be committed:
  (use "git restore --staged <file>..." to unstage)
        modified:   jupiter1.rb


TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git restore --staged jupiter1.rb

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git status
On branch main
Your branch is up to date with 'origin/main'.

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git restore <file>..." to discard changes in working directory)
        modified:   jupiter1.rb

no changes added to commit (use "git add" and/or "git commit -a")

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff
warning: in the working copy of 'jupiter1.rb', LF will be replaced by CRLF the next time Git touches it
diff --git a/jupiter1.rb b/jupiter1.rb
index e69de29..3e27566 100644
--- a/jupiter1.rb
+++ b/jupiter1.rb
@@ -0,0 +1,7 @@
+- name: Setup Tomcat8 & Deploy Artifact
+  hosts: appsrvgrp
+  # gather_facts: no
+  vars:
+    timestamp: "{{ansible_date_time.date}}{{ansible_date_time.hour}}{{ansible_date_time.minute}}"
+    tom_url: https://archive.apache.org/dist/tomcat/tomcat-8/v8.5.37/bin/apache-tomcat-8.5.37.tar.gz
+

5.	If alredy commited the file then how will rollback it ..

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git add .
warning: in the working copy of 'jupiter1.rb', LF will be replaced by CRLF the next time Git touches it

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git commit -m "added jupiter1.rb for rollback practice"
[main a948c07] added jupiter1.rb for rollback practice
 1 file changed, 7 insertions(+)

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git log --oneline
a948c07 (HEAD -> main) added jupiter1.rb for rollback practice
432e995 (origin/sprint1, origin/main, origin/HEAD) jupiter changes
06128c0 Add resource descriptions to moon10.py
6bffce5 updated moons mines name
2051ca2 moon1 mines details
bd336a6 new files commited
dc5cc6e Added Petclinic folder
975ac0b Update MAVEN.md
de8d755 Update MAVEN.md
c9d6076 Create MAVEN.md
c020f62 Added springboot-java-project properly
e13f9f6 Added springboot-java-project
b7c04b0 Update README.md
fdce5fd Update README.md
254233c Update README.md
b75fcc5 Add files via upload
924d0f4 Add files via upload
4d8c58c Update README.md
6f8d882 Update README.md
374cdc3 Update README.md
c2e9f26 Update README.md
6453020 Update README.md
9bac826 Update README.md
e191825 Initial commit

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git status
On branch main
Your branch is ahead of 'origin/main' by 1 commit.
  (use "git push" to publish your local commits)

nothing to commit, working tree clean

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git diff --cached




For that we need to use the cmd is “git log --oneline” to get the commit id ..then copy the commit id …1st we can check the different using the commt id of presnt and previous one “git diff (corrent commit id)..(previous commit id)”…for rollback the recent one “git revert HEAD (commit id)” then provide the revert commit message why want to revert.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git revert HEAD

 After that we need mention revert description like why want to revert
After that we can see the status using “git log - -oneline”

 TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git log --oneline
05b1223 (HEAD -> main) Revert "successfuly  jupiter1.rb using cmd git revert HEAD"
a948c07 added jupiter1.rb for rollback practice
432e995 (origin/sprint1, origin/main, origin/HEAD) jupiter changes
06128c0 Add resource descriptions to moon10.py
6bffce5 updated moons mines name
2051ca2 moon1 mines details
bd336a6 new files commited
dc5cc6e Added Petclinic folder
975ac0b Update MAVEN.md
de8d755 Update MAVEN.md
c9d6076 Create MAVEN.md
c020f62 Added springboot-java-project properly
e13f9f6 Added springboot-java-project
b7c04b0 Update README.md
fdce5fd Update README.md
254233c Update README.md
b75fcc5 Add files via upload
924d0f4 Add files via upload
4d8c58c Update README.md
6f8d882 Update README.md
374cdc3 Update README.md
c2e9f26 Update README.md
6453020 Update README.md
9bac826 Update README.md
e191825 Initial commit

6.	Suppose you don’t want to track ya any one to see the revert status / rollback status then
“git reset - -hard (commit id)“ this is the direct way of the rollback.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git reset --hard 432e995
HEAD is now at 432e995 jupiter changes

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ git log --oneline
432e995 (HEAD -> main, origin/sprint1, origin/main, origin/HEAD) jupiter changes
06128c0 Add resource descriptions to moon10.py
6bffce5 updated moons mines name
2051ca2 moon1 mines details
bd336a6 new files commited
dc5cc6e Added Petclinic folder
975ac0b Update MAVEN.md
de8d755 Update MAVEN.md
c9d6076 Create MAVEN.md
c020f62 Added springboot-java-project properly
e13f9f6 Added springboot-java-project
b7c04b0 Update README.md
fdce5fd Update README.md
254233c Update README.md
b75fcc5 Add files via upload
924d0f4 Add files via upload
4d8c58c Update README.md
6f8d882 Update README.md
374cdc3 Update README.md
c2e9f26 Update README.md
6453020 Update README.md
9bac826 Update README.md
e191825 Initial commit



**************GIT SSH LOGIN*****************

  The authenticate based on usernames and passwords which was not a good idea. Then remember the password and store the password, there are chance of exposing the password.

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ cat .git/config
[core]
        repositoryformatversion = 0
        filemode = false
        bare = false
        logallrefupdates = true
        symlinks = false
        ignorecase = true
[remote "origin"]
        url = https://github.com/Tks-Devops/DevOps-Nursery-to-Graduation.git
        fetch = +refs/heads/*:refs/remotes/origin/*
[branch "main"]
        remote = origin
        merge = refs/heads/main




 So github provides ssh based login and almost every remote git repo provide the ssh based authentication.
For generating the ssh we need tp use this cmd “ssh-keygen.exe” generate the public key and private key ..

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ ssh-keygen.exe
Generating public/private ed25519 key pair.
Enter file in which to save the key (/c/Users/TK Sahu/.ssh/id_ed25519):
Enter passphrase for "/c/Users/TK Sahu/.ssh/id_ed25519" (empty for no passphrase):
Enter same passphrase again:
Your identification has been saved in /c/Users/TK Sahu/.ssh/id_ed25519
Your public key has been saved in /c/Users/TK Sahu/.ssh/id_ed25519.pub
The key fingerprint is:
SHA256:iiXZ7u211g2UyWfkZRT6h/ZGvMFAfz8KgeY5AuyTgys TK Sahu@DESKTOP-R12R5NH
The key's randomart image is:
+--[ED25519 256]--+
|             . oo|
|    .     . . + o|
|     o   o o B +.|
|    ooo o . * Boo|
|   .o=o.S+ o oo==|
|    .=o.. . o..o=|
| E .. o   .. + .o|
|  .  . . .... .. |
|      ..o..      |
+----[SHA256]-----+



Yes, you can search the path, but in your case you don’t really need to “search” randomly—you just need the correct home path 👍
Let me show you both ways:
________________________________________
✅ 1. Direct way (best)
Just run:
echo ~
👉 This will show your home directory, something like:
/c/Users/TK Sahu
Now check:
ls ~/.ssh/
________________________________________
🔍 2. Search manually (if you really want)
You can search for .ssh folder like this:
find /c/Users -type d -name ".ssh" 2>/dev/null
👉 This will scan and show something like:
/c/Users/TK Sahu/.ssh
________________________________________
🧠 Simple understanding
•	.ssh is always inside your home directory
•	Not inside project folders
•	No need for sudo on Windows Git Bash
________________________________________
❗ If nothing found
Then SSH is not created yet → you must run:
ssh-keygen -t ed25519 -C "your_email@example.com"



TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ ls .ssh/
ls: cannot access '.ssh/': No such file or directory

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ sudo ls .ssh/
bash: sudo: command not found

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ echo ~
/c/Users/TK Sahu

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ ls ~/.ssh/
id_ed25519  id_ed25519.pub  known_hosts  known_hosts.old

TK Sahu@DESKTOP-R12R5NH MINGW64 /e/TK SAHU/DevOps-Nursery-to-Graduation (main)
$ cat ~/.ssh/id_ed25519.pub
ssh-ed25519 AAAAC3NzaC1lZDI1NTE5AAAAIB0QN1lOUj2Sv+XHVLulqt0KcgZuxsnfkfTGnMUPZNMx TK Sahu@DESKTOP-R12R5NH

for configure the ssh –-> ogin into the github a/c then go to top right corner settingsssh and GPG keys section click on that click “new ssh key”—give—title & key then click add ssh key ..


After created like this window will apper




***GIT tags, semantic  Versioning  & more******

*what is a semantic versioning ?

Semantic Versioning is a standard way to give versions to your software so anyone can understand what changed just by looking at the version number.
________________________________________
🔢 Format
MAJOR.MINOR.PATCH
Example:
2.5.1
________________________________________
🧠 Meaning of each part
🔴 1. MAJOR version (2.x.x)
•	Increased when you make breaking changes
•	Old code may stop working
👉 Example:
•	API changes
•	Removing features
________________________________________
🟡 2. MINOR version (x.5.x)
•	Increased when you add new features
•	Backward compatible (old code still works)
👉 Example:
•	Add new API endpoint
•	Add new functionality
________________________________________
🟢 3. PATCH version (x.x.1)
•	Increased for bug fixes
•	No new features, no breaking changes
👉 Example:
•	Fix login bug
•	Fix UI issue
________________________________________
📌 Example flow
•	1.0.0 → Initial release
•	1.0.1 → Bug fix
•	1.1.0 → New feature added
•	2.0.0 → Breaking change
________________________________________
🏷️ How it connects with Git tags
You use versions as tags in Git:
git tag v1.0.0
git tag v1.1.0
git tag v2.0.0
Push tags:
git push origin --tags


*********setup github copilot************


🤖 GitHub Copilot Setup Guide
🔹 What is GitHub Copilot?
GitHub Copilot is an AI coding assistant by GitHub and OpenAI that helps you:
•	Write code faster
•	Generate scripts (Docker, YAML, Bash, etc.)
•	Assist in DevOps tasks
________________________________________
🧰 Prerequisites
•	GitHub account (your account: 👉 GitHub)
•	Code editor (recommended: Visual Studio Code)
•	Active internet connection
________________________________________
⚙️ Step 1: Enable Copilot in GitHub
1.	Go to GitHub Settings
2.	Navigate to Copilot
3.	Click Enable GitHub Copilot
👉 Link: https://github.com/settings/copilot
⚠️ Copilot is paid, but:
•	Free trial available
•	Free for students (GitHub Student Pack)
________________________________________
💻 Step 2: Install Copilot in VS Code
1.	Open Visual Studio Code
2.	Go to Extensions (Ctrl + Shift + X)
3.	Search: GitHub Copilot
4.	Install the extension by GitHub
________________________________________
🔐 Step 3: Sign in to GitHub
After installing:
•	Click Sign in to GitHub
•	Authorize in browser
________________________________________
✅ Step 4: Verify Installation
Create a file:
touch test.py
Type:
# create a function to add two numbers
👉 Copilot will automatically suggest:
def add(a, b):
    return a + b
Press:
Tab → Accept suggestion
________________________________________
⚡ How to Use Copilot (Important Commands)
Action	Shortcut
Accept suggestion	Tab
Next suggestion	Alt + ]
Previous suggestion	Alt + [
Trigger manually	Ctrl + Enter
________________________________________
🚀 Real DevOps Examples
1. Dockerfile generation
Type:
# dockerfile for python flask app
👉 Copilot generates full Dockerfile
________________________________________
2. Kubernetes YAML
Type:
# kubernetes deployment for nginx
________________________________________
3. Bash Script
# script to backup logs daily


***************************THANK YOU************************************
```
